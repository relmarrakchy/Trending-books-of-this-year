```python
import requests
from bs4 import BeautifulSoup

url = "---------------------------------------------"

response = requests.get(url)

soup = BeautifulSoup(response.content, 'html.parser')

books = []

for item in soup.select("li.inner-facade"):
  title = item.find("div", class_="container").find("div", class_="list-meta-wrap").find("div", class_="list-meta").find("h2").find("a").get_text()
  if (item.find("div", class_="container").find("div", class_="list-meta-wrap").find("div", class_="list-meta").find("h2", class_="author").find("a")) :
      author = item.find("div", class_="container").find("div", class_="list-meta-wrap").find("div", class_="list-meta").find("h2", class_="author").find("a").get_text()
  else :
      author = item.find("div", class_="container").find("div", class_="list-meta-wrap").find("div", class_="list-meta").find("h2", class_="author").get_text()
  desc_html = item.find("div", class_="container").find("div", class_="list-meta-wrap").find("div", class_="list-meta").find("div", class_="desc").find("span")
  if (desc_html):
    desc = item.find("div", class_="container").find("div", class_="list-meta-wrap").find("div", class_="list-meta").find("div", class_="desc").find("span").get_text()
  else:
    desc = item.find("div", class_="container").find("div", class_="list-meta-wrap").find("div", class_="list-meta").find("div", class_="desc").get_text()
  price = item.find("div", class_="container").find("div", class_="buy").find("div", class_="buy_clmn buy_small").find("h3").find("div").get_text()

  book = {
      "Title": title,
      "Author": author,
      "Description": desc,
      "Price": price
  }

  books.append(book)
```

```python
trending_books = pd.DataFrame(books)

trending_books
```
